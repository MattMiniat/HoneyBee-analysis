```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
file_to_load = "Resources/Honey_State_Production.csv"
file2_to_load = "Resources/National_Honey_Sales.csv"
file3_to_load = "Resources/Honey_US_Production.csv"
file4_to_load = "Resources/Honey_Survey_Prices.csv"
state_production_data = pd.read_csv(file_to_load)
census_price_data = pd.read_csv(file2_to_load)
national_production_data = pd.read_csv(file3_to_load)
survey_price_data = pd.read_csv(file4_to_load)
```

```python
state_honey_df = pd.DataFrame(state_production_data)
national_honey_df = pd.DataFrame(national_production_data)
survey_prices = pd.DataFrame(survey_price_data)
census_prices = pd.DataFrame(census_price_data)
```

# Production Data Texas vs US

### Texas Production Data

```python
state_honey_df["Value"] = pd.to_numeric(state_honey_df["Value"].str.replace(',',''), errors='coerce')
```

```python
Texas_honey = state_honey_df.loc[(state_honey_df["State"] == "TEXAS")]
Texas_production = Texas_honey[["Year" , "Value"]]
Texas_production
```

### US Production Data

```python
national_honey_df["Value"] = pd.to_numeric(national_honey_df["Value"].str.replace(',',''), errors='coerce')
```

```python
US_production = national_honey_df[["Year" , "Value"]]
US_production
```

## Graphs

```python
Texas_production_plot = Texas_production.plot.line(x='Year', y='Value', xticks= Texas_production['Year'])
Texas_production_plot.set_title("Texas Honey Production 1997 - 2017 (USDA Census)")
Texas_production_plot.set_xlabel("Year")
Texas_production_plot.set_ylabel("Production (lbs)")

US_production_plot = US_production.plot.line(x='Year', y='Value', xticks= US_production['Year'])
US_production_plot.set_title("U.S. Honey Production 1997 - 2017 (USDA Census)")
US_production_plot.set_xlabel("Year")
US_production_plot.set_ylabel("Production (per 100,000,000 lbs)")
```

# Sales Data

### United States sales data

```python
National_prices = survey_prices.loc[(survey_prices["State"] == "US TOTAL")]
```

```python
US_Prices = National_prices[["Year","State","Value"]]
```

```python
Real_Prices = round(US_Prices["Value"]/100,2)
US_Prices["Value"] = Real_Prices
```

### Texas sales data

```python
Texas_prices = survey_prices.loc[(survey_prices["State"] == "TEXAS")]
Texas_prices
```

```python
T_Prices = Texas_prices[["Year", "State", "Value"]]
```

```python
Real_Prices = round(T_Prices["Value"]/100,2)
T_Prices["Value"] = Real_Prices
```

## Sales Graphs

```python
T_price_plot = T_Prices.plot.line(x='Year', y='Value')
price_plot = US_Prices.plot.line(x='Year', y='Value')

T_price_plot.set_title("Texas Honey Prices 1987 - 2018 (USDA Survey)")
T_price_plot.set_xlabel("Year")
T_price_plot.set_ylabel("Price per $1 USD")

price_plot.set_title("U.S. Honey Prices 1987 - 2018 (USDA Survey)")
price_plot.set_xlabel("Year")
price_plot.set_ylabel("Price per $1 USD")

plt.show()
```

```python
x = US_Prices['Year']
y1= US_Prices['Value']
y2 = T_Prices['Value']

US, = plt.plot(x,y1, label = "United States")
Texas, = plt.plot(x,y2, label = "Texas")
plt.legend(handles=[US, Texas])
plt.title("Honey sales USD/lb (USDA Survey)")


plt.show()
```
